# Aerofit Business Case

## Business Problem

The market research team at AeroFit wants to identify the characteristics of the target audience for each type of treadmill offered by the company, to provide a better recommendation of the treadmills to the new customers. The team decides to investigate whether there are differences across the product with respect to customer characteristics.

- Perform descriptive analytics to create a customer profile for each AeroFit treadmill product by developing appropriate tables and charts.
- For each AeroFit treadmill product, construct two-way contingency tables and compute all conditional and marginal probabilities along with their insights/impact on the business.

## About the dataset

The given dataset contains 9 columns.

- Product Purchased:	KP281, KP481, or KP781
- Age:	In years
- Gender:	Male/Female
- Education:	In years
- MaritalStatus:	Single or partnered
- Usage:	The average number of times the customer plans to use the treadmill each week.
- Income:	Annual income (in $)
- Fitness:	Self-rated fitness on a 1-to-5 scale, where 1 is the poor shape and 5 is the excellent shape.
- Miles:	The average number of miles the customer expects to walk/run each week

## Product Portfolio

- The KP281 is an entry-level treadmill that sells for  1,500 Dollar.
- The KP481 is for mid-level runners that sell for 1,750 Dollar.
- The KP781 treadmill is having advanced features that sell for 2,500 Dollar.

# Marginal Probablities

The probablity of buying a product for each category.

```python
num_kp281 = len(df.loc[df['Product']=='KP281'])
num_kp481 = len(df.loc[df['Product']=='KP481'])
num_kp781 = len(df.loc[df['Product']=='KP781'])
total_products = df.shape[0]
```

```python
print('Probablity of buying KP281 is ',round(num_kp281/total_products,2))
print('Probablity of buying KP481 is ',round(num_kp481/total_products,2))
print('Probablity of buying KP781 is ',round(num_kp781/total_products,2))
```

# Conditional Probablities

### Probablity of buying a product given it is a male.

```python
gender_tab = pd.crosstab(index = df['Product'],columns = df['Gender'])
gender_tab
```

```python
kp281_male = gender_tab.loc['KP281','Male']
kp481_male = gender_tab.loc['KP481','Male']
kp781_male = gender_tab.loc['KP781','Male']
total_male = gender_tab['Male'].sum()
print('Probablity of buying KP281 given Male is ',round(kp281_male/total_male,2))
print('Probablity of buying KP481 given Male is ',round(kp481_male/total_male,2))
print('Probablity of buying KP781 given Male is ',round(kp781_male/total_male,2))
```

## Probablity of buying a product given female.

```python
kp281_female = gender_tab.loc['KP281','Female']
kp481_female = gender_tab.loc['KP481','Female']
kp781_female = gender_tab.loc['KP781','Female']
total_female = gender_tab['Female'].sum()
print('Probablity of buying KP281 given Female is ',round(kp281_female/total_female,2))
print('Probablity of buying KP481 given Female is ',round(kp481_female/total_female,2))
print('Probablity of buying KP781 given Female is ',round(kp781_female/total_female,2))
```

## Probablity of buying a product given a person is single

```python
maritalstatus_tab  = pd.crosstab(index = df['Product'],columns = df['MaritalStatus'])
maritalstatus_tab
```

```python
kp281_single = maritalstatus_tab.loc['KP281','Single']
kp481_single = maritalstatus_tab.loc['KP481','Single']
kp781_single = maritalstatus_tab.loc['KP781','Single']
total_single = maritalstatus_tab['Single'].sum()
print('Probablity of buying KP281 given Single is',round(kp281_single/total_single,2))
print('Probablity of buying KP481 given Single is',round(kp481_single/total_single,2))
print('Probablity of buying KP781 given Single is',round(kp781_single/total_single,2))
```

## Probablity of buying a product given a person is married

```python
kp281_married = maritalstatus_tab.loc['KP281','Partnered']
kp481_married = maritalstatus_tab.loc['KP481','Partnered']
kp781_married = maritalstatus_tab.loc['KP781','Partnered']
total_married = maritalstatus_tab['Partnered'].sum()
print('Probablity of buying KP281 given Partnered is ',
      round(kp281_married/total_married,2))
print('Probablity of buying KP481 given Partnered is ',
      round(kp481_married/total_married,2))
print('Probablity of buying KP781 given Partnered is ',
      round(kp781_married/total_married,2))
```

## Recommendations and Actionable Insights

- Based on the above customer profiling we can target specific customers groups of specific products.
- Based on income rangees we can target partnered customers and male customers are more likely to buy:
    1. We observed, median income of partnered customers is more than single customers, so partnered customers are more likely to buy.
    2. We observed, median income of male buyers is more than female buyers, so male customers are more likely to buy.
    3. For KP281, the income ranges of customers are from 25000-50000 dollars and for KP481 the income ranges of cutomers are from 25000-75000 dollars and for KP781 the income ranges of customers are from 75000-100000 dollars,so accordingly we can target the customers based on their income range.

- Based on usage, it is observed the customers who expect to use it for 2-3 times buy KP281 and customers who expect it to use 3-4 times or more per week buy KP781. So we can suggest KP281 to customers who want to use it for less than 3 times and KP781 to customers who want to use it for more than 3 times.

- Based on fitness score, we can suggest KP781 to customers who have a fitness score >= 3.0.

- Based on number of miles expected to  walk/run each week , we can suggest KP781 to customers who are expecting to walk/run more than the median of miles expected to walk/run of other two products.
